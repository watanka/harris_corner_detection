```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
# load image
imgfile = 'house_corners.png'
img = plt.imread(imgfile)
```

```python
# get gradients

x_filter = np.array([[-1,0,1],
                     [-2,0,2],
                     [-1,0,1]])
y_filter = np.array([[1,2,1],
                     [0,0,0],
                     [-1,-2,-1]])
```

```python
x_grad = cv2.filter2D(img, -1, x_filter)
y_grad = cv2.filter2D(img, -1, y_filter)
```

```python
plt.imshow(x_grad)
```

```python
plt.imshow(y_grad)
```

```python
plt.scatter(x_grad, y_grad)
```
